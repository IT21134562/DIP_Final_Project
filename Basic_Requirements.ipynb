```python
from google.colab import files
from PIL import Image
import IPython.display as display
from PIL import Image
import matplotlib.pyplot as plt
import PIL.ImageOps

uploaded_image = None

def upload_image():
    global uploaded_image
    uploaded = files.upload()
    for name, data in uploaded.items():
        uploaded_image = Image.open(name)
        display.display(uploaded_image)
        return uploaded_image

uploaded_image = upload_image()


#Color Change Image
if uploaded_image is not None:
    bw_image = uploaded_image.convert("1")
    grayscale_image = uploaded_image.convert("L")

    #Color change images displaying
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(uploaded_image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(bw_image)
    axes[1].set_title('Black and White Image')
    axes[1].axis('off')

    axes[2].imshow(grayscale_image, cmap='gray')
    axes[2].set_title('Grayscale Image')
    axes[2].axis('off')
    plt.show()

print("")
pass

#Rotate Image
if uploaded_image is not None:
    angle = int(input("Enter the rotation angle (in degrees): "))

    # Rotate the image
    rotated_image = uploaded_image.rotate(angle)

    #Rotated image displying
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(uploaded_image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(rotated_image)
    axes[1].set_title('Rotated Image')
    axes[1].axis('off')
    plt.show()

print("")
pass


#Image Crop
if uploaded_image is not None:
    left = int(input("Enter the left cropping dimension: "))
    top = int(input("Enter the top cropping dimension: "))
    right = int(input("Enter the right cropping dimension: "))
    bottom = int(input("Enter the bottom cropping dimension: "))

    #Cropped image displying
    angle = uploaded_image.crop((left, top, right, bottom))

    # display
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(uploaded_image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(angle)
    axes[1].set_title('Cropped Image')
    axes[1].axis('off')
    plt.show()

```
