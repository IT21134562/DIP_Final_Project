```python
from importlib.metadata import files
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
from PIL import Image
import IPython.display as display
from PIL import Image
import PIL.ImageOps

# Load the image 
image = cv2.imread('D:/depositphotos_54081723-stock-photo-beautiful-nature-landscape.jpg')


#Color Invert

#Original image
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 
  
# Inverse by subtracting from 255 
inverse_image = 255 - image 
  
#Inverse image 
plt.subplot(1, 2, 2) 
plt.title("Inverse color") 
plt.imshow(inverse_image) 
plt.show()
```
